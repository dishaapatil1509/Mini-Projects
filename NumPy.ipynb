```python
import pandas as pd
import numpy as np
```

```python
#Generate marks for 50 students.

np.random.randint(1,100,51)
```

```python
#Calculate total, average, highest, lowest.

marks = np.random.randint(0, 101, 50)

print("Marks:", marks)
print("Total:", np.sum(marks))
print("Average:", np.mean(marks))
print("Highest:", np.max(marks))
print("Lowest:", np.min(marks))
```

```python
#Count above 75, below 35, between 35-75.

marks = np.random.randint(0,101,50)

print("Above 75:", np.sum(marks > 75))
print("Below 35:", np.sum(marks < 35))
print("35 to 75:", np.sum((marks >= 35) & (marks <= 75)))
```

```python
#Assign grades.

marks = np.array([95,82,75,68,55,30])
grade = []

for i in marks:
    if i>=90:
        grade.append("A")
    elif i>=75:
        grade.append("B")
    elif i>=60:
        grade.append("C")
    elif i>=35:
        grade.append("D")
    else:
        grade.append("Fail")
print(marks)
print(grade)
```

```python
#Display top 5 marks.

marks = np.random.randint(0, 101, 50)
marks.sort()
print("Top 5 Marks:")
print(marks[-5:])
```

```python
#Display bottom 5 marks.

marks = np.random.randint(0, 101, 50)
marks.sort()
print("Bottom 5 Marks:")
print(marks[:5])
```

```python
#Find median and standard deviation

marks = np.random.randint(0, 101, 50)

print("Median:", np.median(marks))
print("Standard Deviation:", np.std(marks))
```

```python
#Create a 5-subject marks matrix and calculate totals

marks = np.random.randint(35, 101, (5,5))

print("Marks Matrix:")
print(marks)
total = np.sum(marks, axis=1)
print("Total Marks:",total)
```

```python
#Find topper in each subject

marks = np.random.randint(35, 101, (5,5))

print("Marks Matrix:")
print(marks)
print("Highest marks in each subject:",np.max(marks, axis=0))
```

```python
#Generate report card with Total, Average, Percentage and Grade

marks = np.random.randint(35, 101, (5,5))

print("Marks:")
print(marks)

total = np.sum(marks, axis=1)
average = np.mean(marks, axis=1)
percentage = total / 500 * 100

print("\nReport Card")

for i in range(5):
    if percentage[i] >= 80:
        grade = "A"
    elif percentage[i] >= 60:
        grade = "B"
    elif percentage[i] >= 40:
        grade = "C"
    else:
        grade = "Fail"

    print("Student", i + 1)
    print("Total:", total[i])
    print("Average:", average[i])
    print("Percentage:", percentage[i])
    print("Grade:", grade)
    print()
```
