```python
#Create a student record system that allows adding student data (name, roll, marks) and saving the data to a file.

f = open("Student_details.txt","w")

for i in range(4):
    name = input("Enter ur name:")
    rno = input("Enter ur roll no:")
    marks = input("Enter ur marks:")
    f.write(name + " , " + rno + " , " + marks + "\n")

f.close()
```

```python
#Implement search functionality to find a student by roll number.

f=open("Student_details.txt","r")
roll=input("Enter your Roll Number to search:")
found=False

for i in f:
    if roll in i:
        print("Record Found")
        print(i)
        found=True

if found==False:
    print("Record not found")
```

```python
#Add functionality to update student marks.

roll = input("Enter roll no: ")
new_marks = input("Enter new marks: ")

f = open("Student_details.txt", "r")
lines = f.readlines()
f.close()

f = open("Student_details.txt", "w")

for i in lines:
    data = i.strip().split(",")

    if data[1].replace("Roll no is:", "").strip() == roll:
        data[2] = " Marks: " + new_marks
        i = ",".join(data) + "\n"

    f.write(i)

f.close()

print("Record Updated")
```

```python
#Delete a student record by name or roll number.

value = input("Enter name or roll no: ")

f = open("Student_details.txt", "r")
lines = f.readlines()
f.close()

f = open("Student_details.txt", "w")

for i in lines:
    if value not in i:
        f.write(i)

f.close()

print("Record Deleted")
```

```python
#Show all student records sorted by marks.

f = open("Student_details.txt", "r")

students = []

for i in f:
    data = i.strip().split(",")

    marks = int(data[2].replace("Marks:", "").strip())

    students.append([data[0], data[1], marks])

students.sort(key=lambda x: x[2])

for i in students:
    print(i)

f.close()
```

```python
#Display student records with marks greater than 75.

f = open("Student_details.txt", "r")

for i in f:
    data = i.strip().split(",")

    marks = int(data[2].replace("Marks:", "").strip())

    if marks > 75:
        print(i)

f.close()
```

```python
#Export top 3 performing students to a new file.

f = open("Student_details.txt", "r")

students = []

for i in f:
    data = i.strip().split(",")
    marks = int(data[2].replace("Marks:", "").strip())
    students.append((marks, i))
f.close()
students.sort(reverse=True)
top = open("Top3.txt", "w")

for i in range(3):
    top.write(students[i][1])

top.close()
print("Top 3 students exported successfully.")
```

```python
#Implement data backup feature (copying all data to backup.txt).

f = open("Student_details.txt", "r")
data = f.read()
f.close()

backup = open("backup.txt", "w")
backup.write(data)
backup.close()

print("Backup created successfully.")
```

```python
#Count how many students scored more than 90 and less than 35.

f = open("Student_details.txt", "r")

above90 = 0
below35 = 0

for i in f:
    data = i.strip().split(",")

    marks = int(data[2].replace("Marks:", "").strip())

    if marks > 90:
        above90 += 1

    if marks < 35:
        below35 += 1

f.close()

print("Students scoring more than 90:", above90)
print("Students scoring less than 35:", below35)
```

```python
# Create a menu-driven system (Add, View, Search, Update, Delete, Exit).

while True:

    print("\n----- MENU -----")
    print("1. Add")
    print("2. View")
    print("3. Search")
    print("4. Update")
    print("5. Delete")
    print("6. Exit")

    ch = int(input("Enter your choice: "))

    if ch == 1:
        f = open("Student_details.txt", "a")
        name = input("Enter name: ")
        roll = input("Enter roll no: ")
        marks = input("Enter marks: ")
        f.write("Name is: " + name + " , Roll no is: " + roll + " , Marks: " + marks + "\n")
        f.close()
        print("Record Added")

    elif ch == 2:
        f = open("Student_details.txt", "r")
        for i in f:
            print(i)
        f.close()

    elif ch == 3:
        roll = input("Enter roll no: ")
        f = open("Student_details.txt", "r")
        found = False
        for i in f:
            if roll in i:
                print("Record Found")
                print(i)
                found = True
        if found == False:
            print("Record Not Found")
        f.close()

    elif ch == 4:
        roll = input("Enter roll no: ")
        new_marks = input("Enter new marks: ")
        f = open("Student_details.txt", "r")
        lines = f.readlines()
        f.close()
        f = open("Student_details.txt", "w")
        for i in lines:
            data = i.strip().split(",")
            if data[1].replace("Roll no is:", "").strip() == roll:
                data[2] = " Marks: " + new_marks
                i = ",".join(data) + "\n"

            f.write(i)

        f.close()
        print("Record Updated")

    elif ch == 5:
        value = input("Enter name or roll no: ")

        f = open("Student_details.txt", "r")
        lines = f.readlines()
        f.close()

        f = open("Student_details.txt", "w")

        for i in lines:
            if value not in i:
                f.write(i)

        f.close()
        print("Record Deleted")

    elif ch == 6:
        print("Thank You!")
        break

    else:
        print("Invalid Choice")
```
